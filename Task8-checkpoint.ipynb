# importing libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import seaborn as sns
```

```python
df = pd.read_csv("Mall_Customers.csv")
df.head()

```

```python
print(df.info())
```

# PCA for 2D visualization

```python
numeric_df = df.select_dtypes(include=['number'])

# Now apply PCA on numeric data only
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(numeric_df)

plt.figure(figsize=(8, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], s=30)
plt.title('PCA - Raw Data Visualization (Numeric Only)')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

# Fit K-Means and assign cluster labels

```python
from sklearn.cluster import KMeans

# Use numeric-only data
numeric_df = df.select_dtypes(include=['number'])

# Fit KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(numeric_df)

# View the assigned clusters
print(df[['Cluster']].head())
```

#  Use Elbow Method to Find Optimal K

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import os
os.environ["OMP_NUM_THREADS"] = "1"

# Select numeric features
numeric_df = df.select_dtypes(include=['number'])

sse = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(numeric_df)
    sse.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K_range, sse, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('SSE (Inertia)')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Step 1: Encode or use numeric features
data = pd.get_dummies(df)  # or use df.select_dtypes(include='number')

# Step 2: Reduce dimensions for visualization
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(data)

# Step 3: Fit KMeans with chosen K
kmeans = KMeans(n_clusters=3, random_state=42)
cluster_labels = kmeans.fit_predict(data)

# Step 4: Visualize
plt.figure(figsize=(8, 6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1],
            c=cluster_labels, cmap='viridis', s=50)

# Optional: plot cluster centers (also projected to PCA space)
centers = pca.transform(kmeans.cluster_centers_)
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centroids')

plt.title("KMeans Clustering Visualization (2D PCA)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.legend()
plt.grid(True)
plt.show()
```
